## 2. Stochastic Gradient Descent (SGD)

## WHY SHUFFLING IS NEEDED IN SGD
Your dataset is 506 dishes (houses). SGD means you taste just one dish at a time, adjust your recipe (weights), and move to the next dish.

Here is what happens WITHOUT shuffling (your code) vs. WITH shuffling:

# The "No Shuffle" Disaster (Sorted by Price)
Imagine your dataset is perfectly sorted from Cheapest houses ($50k)** to **Most Expensive houses ($500k).

- Epoch 1, Dish 1–100 (Cheap): You taste cheap food. You adjust your recipe to make perfect cheap food. By the time you reach Dish 100, your recipe is excellent for cheap food.

- Epoch 1, Dish 101–200 (Medium): You taste medium-priced food. Your recipe is completely wrong for it. You adjust the recipe. Now your recipe works for medium food, but... you have completely forgotten how to cook cheap food!

- Epoch 1, Dish 300–506 (Expensive): You adjust the recipe to cook expensive food. Now you are great at expensive food, but forgot cheap and medium.

- End of Epoch 1: Your recipe only works for expensive houses.
Start of Epoch 2: You start again at Dish 1 (Cheap). You adjust the recipe back to cheap food. You forget expensive food again.

Your weights just oscillate violently back and forth like a pendulum. The recipe never settles on the "average" cooking style that works for all prices.

# Why Shuffling Saves You --
Now imagine you shuffle the menu before each epoch. You taste Dish 1 (Cheap), then Dish 2 (Expensive), then Dish 3 (Medium), then Dish 4 (Cheap), etc.

- Update 1 (Cheap): Recipe adjusts slightly towards cheap food.

- Update 2 (Expensive): Recipe adjusts back slightly towards expensive food.

- Update 3 (Medium): Recipe adjusts back towards medium food.

Because the cheap, medium, and expensive dishes are mixed up randomly, the updates cancel each other's extreme directions out and average out to point straight downhill toward the middle.

The weights take a steady, wobbly walk toward the best overall recipe, rather than violently swinging from one extreme to the other.

# The Mathematical Reason (The "Unbiased Estimate")
In Batch GD, you use all 506 houses to calculate the gradient. This is the exact downhill direction.

In SGD, you use just 1 house to estimate the downhill direction.

For this estimate to be correct on average (mathematically "unbiased"), that 1 house must be chosen randomly.

If you pick them in a fixed sorted order:

Your estimates are biased for long stretches (cheap bias for 100 steps, expensive bias for 100 steps).

The math assumes E[gradient_SGD] = gradient_Batch (the average of your single-house gradients equals the true gradient). This assumption breaks entirely if the data has any hidden order.

# Human Analogy (The "Study" Test)
No Shuffle (Sequential): Study Math for 5 hours straight. Then Physics for 5 hours. Then History for 5 hours. At the end of the day, you only remember History. Your brain oscillates.

Shuffle (Random): Study 30 mins of Math, 30 mins of Physics, 30 mins of History, then back to Math. Your brain constantly refreshes all subjects and learns them simultaneously. This is exactly what shuffling does for your model!

#T he ONLY Exception (When you don't need to shuffle)
You only don't need to shuffle if your dataset is already perfectly random (e.g., the rows are completely shuffled by default). In real life, this never happens. Datasets are often sorted by date, price, ID, or location—so always shuffle in SGD/Mini-Batch!

# Quick Visual Proof
If you run your code with epoch2 = 500, learning_rate2 = 0.0001, but remove the shuffling, you will see:

The loss graph looks like a sawtooth (down then up, down then up) with huge spikes.

After 500 epochs, the final loss is still high (~100+).

If you add shuffling, the loss graph becomes a smooth curve, and the final loss drops low (~20).

That one line of code (np.random.permutation) is the difference between failure and success! 🚀

## 3. MINI-BATCH GRADIENT

```python
epoch3 = 5000
weights3 = np.random.randn(X_final.shape[1],1)*0.01

learning_rate3 = 0.01
batch_size = 50

loss_hist3 = []

for i in range(0,epoch3):

    shuffle_idx = np.random.permutation(N)
    X_jj = X_final[shuffle_idx]
    y_jj = y[shuffle_idx]

    for j in range(0,N,batch_size):
        x_j = X_jj[j:j+batch_size]
        y_j = y_jj[j:j+batch_size]

        y_pred3 = x_j @ weights3

        error3 = y_pred3 - y_j

        gradient3 = (2/x_j.shape[0]) *(x_j.T @ error3)

        weights3 = weights3 - learning_rate3 * gradient3

    y_predAll3 = X_final @ weights3
    loss3 = np.mean((y_predAll3 - y)**2)
    loss_hist3.append(loss3)
    
    if(i%50==0):
        print(f"loss at epoch {i} is {loss3}")

print("\nTraing complete")
print(f"wights are : {weights3.flatten()}")

# PLOT
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(loss_hist3)
plt.title(" Loss over Epochs")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.scatter(X_scaled[:, 0], y, alpha=0.5, label="Actual")  # Plotting RM (index 5) is better, but using feature 0 (CRIM) for demonstration
plt.scatter(X_scaled[:, 0], y_predAll3, alpha=0.5, label="Predicted", color="orange")
plt.xlabel("Scaled Feature")
plt.ylabel("Price")
plt.legend()
plt.show()
```

## MAE LOSS

```python
def gradMAE(error,num,Xf):
    grad = (1/num) * (Xf.T @ np.sign(error))
    return grad

def lossMAE(error):
    return np.abs(np.mean(error))
```

## HUBER LOSS

```python
def gradHuber(error,num, Xf,delta):
    gradError = np.where(np.abs(error)<=delta, error, np.sign(error))
    grad = (1/num)*(Xf.T @ gradError)
    return grad

def lossHuber(error, delta):
    loss_temp = np.where(np.abs(error) <= delta, 0.5*error**2, delta*(np.abs(error) - 0.5*delta))
    return np.mean(loss_temp)
```
